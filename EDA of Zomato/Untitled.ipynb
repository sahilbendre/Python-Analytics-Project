```python
plt.figure(figsize=(12,5))
plt.bar(list(df["Restaurant Name"].value_counts()[0:5].keys()),list(df["Restaurant Name"].value_counts()[0:5]))
```

##conclution 

in this chart we see cafe coffee day, domino's pizza, subway, green chick chop, McDonald's have many order from this store all over the india.

```python
plt.figure(figsize=(18,8))
plt.bar(list(df["City"].value_counts()[0:10].keys()),list(df["City"].value_counts()[0:10]))
```

here we are see New Delhi people have more order. and after that gurgaon, noida people have more order.

```python
#making Pie Char
plt.figure(figsize=(20,15))
plt.pie(list(df["City"].value_counts()),labels=list(df["City"].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
