```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.head()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df.describe()
```

## Exploratory Data Analysis

###### Gender

```python
ax=sns.countplot(x="Gender",data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

    From Above Graph we can see that most of the buyer as Female and even the purchasing power of Females are Greater than Male.

###### Age

```python
df.columns
```

```python
ax=sns.countplot(x="Age Group",data=df,hue="Gender")
for bar in ax.containers:
    ax.bar_label(bar)
    
```

```python
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From Above Graph We can See that most of the Buyer Are of  Age Group between 26-35 yrs.

###### state

```python
sales_states=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x="State",y="Orders",data=sales_states)
```

```python
sales_states=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x="State",y="Amount",data=sales_states)
```

From Above graphs we can see the most of the orders & total sales amount from Uttar Pradesh, Maharashtra And Karnataka respectivly

###### Marital Status

```python
sns.set(rc={'figure.figsize':(7,5)})
ax=sns.countplot(x="Marital_Status",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Marital_Status",y="Amount",data=sales_state,hue="Gender")
```

From above graph we can see that most of the Buyer are married Women and they have high purchasing power

###### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x="Occupation",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x="Occupation",y="Amount",data=sales_states)
```

From Above Graph we can see that most of the buyers are Working in IT, Healthcare and Aviation Sector.

###### Products

```python
sns.set(rc={'figure.figsize':(25,6)})
ax=sns.countplot(x="Product_Category",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x="Product_Category",y="Amount",data=sales_states)
```

From Above Graph we can see the most of the sold products are Food, clothing, Electronics Category
