```python
def categorized_age(age):
    if age<=12:
        return "children"
    elif age>=13 and age<=19:
        return "Teens"
    elif age>=20 and age<=39:
        return "Adult"
    elif age>=40 and age<=59:
        return "Middle age adults"
    elif age>60:
        return "Senior Citizen"
df['Age Group']=df["Age"].apply(categorized_age)
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.color_palette("rocket")
ax=sns.countplot(x=df["Age Group"])
for bars in ax. containers:
    ax.bar_label(bars)
plt.title("Distribution of Age Groups by Count", fontsize=14,fontweight="bold")
```

This bar chart represents the count of different age groups, including **Teens, Adults, Middle Age Adults, and Senior Citizens**. The data shows that **Middle Age Adults** have the highest count, followed by **Adults**, while **Teens** have the lowest count.

### Observations:
- **Middle Age Adults**: 1510 (Highest)
- **Adults**: 1452
- **Senior Citizens**: 723
- **Teens**: 150 (Lowest)
